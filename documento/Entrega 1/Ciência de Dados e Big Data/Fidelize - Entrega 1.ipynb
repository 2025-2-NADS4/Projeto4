```python
# GRÁFICO 3: Contagem de Pedidos por Canal de Venda

# Frequência de cada canal de venda
contagem_canais = dataframes['orders']['salesChannel'].value_counts()

plt.figure(figsize=(10, 6))

plt.bar(
  contagem_canais.index, # Eixo X: nomes dos canais
  contagem_canais.values, # Eixo Y: valores da contagem
  color='royalblue'
  )

plt.title('Contagem de Pedidos por Canal de Venda', fontsize=16)
plt.xlabel('Canal de Venda', fontsize=12)
plt.ylabel('Número de Pedidos', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# GRÁFICO 4: Análise do Valor Total do Pedido

plt.figure(figsize=(10, 6))

plt.hist(
  dataframes['orders']['totalAmount'],
  bins=30,
  edgecolor='black',
  alpha=0.75
)

# Média
plt.axvline(
  x=dataframes['orders']['totalAmount'].mean(),
  color='red',
  linestyle='--',
  linewidth=2,
  label=f"Média: R$ {dataframes['orders']['totalAmount'].mean():.2f}"
)

# Mediana
plt.axvline(
  x=dataframes['orders']['totalAmount'].median(),
  color='green',
  linestyle='--',
  linewidth=2,
  label=f"Mediana: R$ {dataframes['orders']['totalAmount'].median():.2f}"
)

plt.title('Distribuição do Valor dos Pedidos', fontsize=16)
plt.xlabel('Valor do Pedido (R$)', fontsize=12)
plt.ylabel('Nº de Pedidos', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# GRÁFICO 5: Valor Médio do Pedido por Tipo

media_por_tipo = dataframes['orders'].groupby('orderType')['totalAmount'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))

bars = plt.bar(
  media_por_tipo.index,
  media_por_tipo.values,
  color=['#2E8B57', '#4682B4', '#CD5C5C']
)

for bar in bars:
  val = bar.get_height()
  plt.text(bar.get_x() + bar.get_width()/2.0, val, f'R$ {val:.2f}', va='bottom', ha='center')

plt.title('Valor Médio do Pedido por Tipo', fontsize=16)
plt.xlabel('Tipo de Pedido', fontsize=12)
plt.ylabel('Valor Médio do Pedido (R$)', fontsize=12)
plt.ylim(0, media_por_tipo.values.max() * 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# GRÁFICO 7: Quantidade de Pedidos + Receita por dia
if not df_orders_cleaned.empty:
    ordem_dias = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    pedidos_por_dia = df_orders_cleaned.groupby('day_of_week')['totalAmount_num'].count().reindex(ordem_dias)
    faturamento_por_dia = df_orders_cleaned.groupby('day_of_week')['totalAmount_num'].sum().reindex(ordem_dias)

    fig, ax1 = plt.subplots(figsize=(14, 7))
    x = np.arange(len(ordem_dias))
    width = 0.3

    color_ax1 = 'tab:blue'
    rects1 = ax1.bar(x - width/2, pedidos_por_dia.values, width, label='Quantidade de Pedidos', color=color_ax1)
    ax1.set_xlabel('Dia da Semana', fontsize=12, weight='bold')
    ax1.set_ylabel('Quantidade de Pedidos', color=color_ax1, fontsize=12)
    ax1.tick_params(axis='y', labelcolor=color_ax1)
    ax1.set_xticks(x)
    ax1.set_xticklabels(ordem_dias)

    ax2 = ax1.twinx()
    color_ax2 = 'tab:green'
    rects2 = ax2.bar(x + width/2, faturamento_por_dia.values, width, label='Receita (R$)', color=color_ax2)
    ax2.set_ylabel('Receita (R$)', color=color_ax2, fontsize=12)
    ax2.tick_params(axis='y', labelcolor=color_ax2)

    # Formatação de moeda
    formatter = mticker.FuncFormatter(lambda x, p: f'R$ {x:,.0f}'.replace(',', '.'))
    ax2.yaxis.set_major_formatter(formatter)

    plt.title('Quantidade de Pedidos vs. Receita por Dia da Semana', fontsize=16)
    fig.tight_layout()
    fig.legend(loc='lower left', bbox_to_anchor=(0.1, 0.9))
    plt.show()
```
