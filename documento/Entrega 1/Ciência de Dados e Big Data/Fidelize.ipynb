```python
import pandas as pd
import io
import matplotlib.pyplot as plt

arquivos_csv = {
    'customers': 'Customer_semicolon.csv',
    'campaigns': 'Campaign_semicolon.csv',
    'orders': 'Order_semicolon.csv',
    'campaign_sends': 'CampaignQueue_semicolon.csv'
}

dataframes = {}

for nome, arquivo in arquivos_csv.items():
        df = pd.read_csv(arquivo, sep=';')
        dataframes[nome] = df
```

```python
# ENTENDENDO OS DADOS

# Tabela de Clientes

df_customers = dataframes['customers']

# Amostra de dados

print("\nAmostra dos dados:")
display(df_customers.head())

# Resumo técnico

print("\nResumo técnico:")
buffer = io.StringIO()
df_customers.info(buf=buffer)
print(buffer.getvalue())
```

```python
# Tabela de Pedidos

df_orders = dataframes['orders']

# Amostra de dados

print("\nAmostra dos dados:")
display(df_orders.head())

# Resumo técnico

print("\n[DESCRIÇÃO] Resumo técnico:")
buffer = io.StringIO()
df_orders.info(buf=buffer)
print(buffer.getvalue())
```

```python
# Tabela de Campanhas

df_campaigns = dataframes['campaigns']

# Amostra de dados
print("\nAmostra dos dados:")
display(df_campaigns.head())

# Resumo técnico
print("\n[DESCRIÇÃO] Resumo técnico:")
buffer = io.StringIO()
df_campaigns.info(buf=buffer)
print(buffer.getvalue())
```

```python
# Tabela de Envio de Campanhas

df_sends = dataframes['campaign_sends']

# Amostra de dados
print("\nAmostra dos dados:")
display(df_sends.head())

# Resumo técnico
print("\n[DESCRIÇÃO] Resumo técnico:")
buffer = io.StringIO()
df_sends.info(buf=buffer)
print(buffer.getvalue())
```

```python
# VERIFICANDO A QUALIDADE DO DADO

# Tabela de Clientes

# Verificação de valores nulos/vazios
print("\nVerificação de valores nulos por coluna:")
print(df_customers.isnull().sum())

# Verificação de linhas duplicadas
num_duplicados = df_customers.duplicated().sum()
print(f"\nNúmero de linhas duplicadas: {num_duplicados}")
```

```python
# Tabela de Pedidos

# Verificação de valores nulos/vazios
print("\n[QUALIDADE] Verificação de valores nulos por coluna:")
print(df_orders.isnull().sum())

# Verificação de linhas duplicadas
num_duplicados = df_orders.duplicated().sum()
print(f"\nNúmero de linhas duplicadas: {num_duplicados}")
```

```python
# Tabela de Campanhas

# Verificação de valores nulos/vazios
print("\nVerificação de valores nulos:")
print(df_campaigns.isnull().sum())

# Verificação de linhas duplicadas
num_duplicados = df_campaigns.duplicated().sum()
print(f"\nNúmero de linhas duplicadas: {num_duplicados}")
```

```python
# Tabela de Envios de Campanha

# Verificação de valores nulos/vazios
print("\nVerificação de valores nulos:")
print(df_sends.isnull().sum())

# Verificação de linhas duplicadas
num_duplicados = df_sends.duplicated().sum()
print(f"\nNúmero de linhas duplicadas: {num_duplicados}")
```

```python
# EXPLORANDO OS DADOS

# Tabela de Clientes

# Contar status dos clientes (1 = Ativo, 2 = Inativo)
print("\nContagem de clientes por status:")
print(df_customers['status'].value_counts())
```

```python
# GRÁFICO 1: Situação dos Clientes
contagem_status = dataframes['customers']['status'].value_counts()

mapa_status = {1: 'Ativo', 2: 'Inativo'}
rotulos = contagem_status.index.map(lambda s: mapa_status.get(s, s))

plt.figure(figsize=(5, 5))

plt.pie(
  contagem_status.values,
  labels=rotulos,
  autopct='%1.1f%%',
  startangle=90,
  colors=['#66b3ff', '#ff9999'],
  explode=(0.05, 0)
)

plt.title('Proporção de Clientes por Status', fontsize=16)
plt.axis('equal')
plt.show()
```

```python
# Tabela de Pedidos

# Principais canais de venda
print("\nContagem de pedidos por canal de venda:")
print(df_orders['salesChannel'].value_counts())

print()

# Tipos de pedido
print("\nContagem de pedidos por tipo:")
print(df_orders['orderType'].value_counts())

print()

# Análise do total do pedido
print("\nAnálise da coluna 'totalAmount':")
display(df_orders['totalAmount'].describe())
```

```python
# GRÁFICO 2: Contagem de Pedidos por Canal de Venda

# Frequência de cada canal de venda
contagem_canais = dataframes['orders']['salesChannel'].value_counts()

plt.figure(figsize=(10, 6))

plt.bar(
  contagem_canais.index, # Eixo X: nomes dos canais
  contagem_canais.values, # Eixo Y: valores da contagem
  color='royalblue'
  )

plt.title('Contagem de Pedidos por Canal de Venda', fontsize=16)
plt.xlabel('Canal de Venda', fontsize=12)
plt.ylabel('Número de Pedidos', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# GRÁFICO 3: Análise do Valor Total do Pedido

plt.figure(figsize=(10, 6))

plt.hist(
  dataframes['orders']['totalAmount'],
  bins=30,
  edgecolor='black',
  alpha=0.75
)

# Média
plt.axvline(
  x=dataframes['orders']['totalAmount'].mean(),
  color='red',
  linestyle='--',
  linewidth=2,
  label=f"Média: R$ {dataframes['orders']['totalAmount'].mean():.2f}"
)

# Mediana
plt.axvline(
  x=dataframes['orders']['totalAmount'].median(),
  color='green',
  linestyle='--',
  linewidth=2,
  label=f"Mediana: R$ {dataframes['orders']['totalAmount'].median():.2f}"
)

plt.title('Distribuição do Valor dos Pedidos', fontsize=16)
plt.xlabel('Valor do Pedido (R$)', fontsize=12)
plt.ylabel('Nº de Pedidos', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# GRÁFICO 4: Valor Médio do Pedido por Tipo

media_por_tipo = dataframes['orders'].groupby('orderType')['totalAmount'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))

bars = plt.bar(
  media_por_tipo.index,
  media_por_tipo.values,
  color=['#2E8B57', '#4682B4', '#CD5C5C']
)

for bar in bars:
  val = bar.get_height()
  plt.text(bar.get_x() + bar.get_width()/2.0, val, f'R$ {val:.2f}', va='bottom', ha='center')

plt.title('Valor Médio do Pedido por Tipo', fontsize=16)
plt.xlabel('Tipo de Pedido', fontsize=12)
plt.ylabel('Valor Médio do Pedido (R$)', fontsize=12)
plt.ylim(0, media_por_tipo.values.max() * 1.1) # Aumenta o limite do eixo Y para o texto caber
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Tabela de Campanhas

# Contagem de campanhas por tipo
print("\nContagem de campanhas por tipo (1 = Promocional, 2 = Institucional):")
print(df_campaigns['type'].value_counts())

print()

# Contagem de campanhas por status
print("\nContagem de campanhas por status (1 = Rascunho, 2 = Cancelado, 3 = Publicado e 4 = Completado):")
print(df_campaigns['status'].value_counts())
```

```python
# GRÁFICO 5: Status das Campanhas

contagem_status_campanha = dataframes['campaigns']['status'].value_counts()

mapa_status_campanha = {1: 'Rascunho', 2: 'Cancelado', 3: 'Publicado', 4: 'Completado'}
rotulos = contagem_status_campanha.index.map(lambda s: mapa_status_campanha.get(s, f'Status {s}'))

plt.figure(figsize=(10, 6))
plt.bar(rotulos, contagem_status_campanha.values, color='#4682B4')

plt.title('Contagem de Campanhas por Status', fontsize=16)
plt.xlabel('Status da Campanha', fontsize=12)
plt.ylabel('Número de Campanhas', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Tabela de Envios de Campanha

# Contagem por status
print("\nContagem de envios por status (1 = Agendado, 2 = Enviado, 3 = Recebido, 4 = Lido, 5 = Deletado, 6 = Pendente):")
print(df_sends['status'].value_counts())
```

```python
# GRÁFICO 6: Status dos Envios de Mensagens

contagem_status_envio = dataframes['campaign_sends']['status'].value_counts()

mapa_status_envio = {1: 'Agendado', 2: 'Enviado', 3: 'Recebido', 4: 'Lido', 5: 'Deletado', 6: 'Pendente'}
rotulos = contagem_status_envio.index.map(lambda s: mapa_status_envio.get(s, f'Status {s}'))

plt.figure(figsize=(10, 6))
plt.bar(rotulos, contagem_status_envio.values, color='#8FBC8F')

plt.title('Status dos Envios de Mensagens', fontsize=16)
plt.xlabel('Status do Envio', fontsize=12)
plt.ylabel('Número de Mensagens', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
